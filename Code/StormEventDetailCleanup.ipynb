```python
# import dependencies
import pandas as pd
import os
```

```python
directory = "Severe Weather Event Data\details"
appended_data = []

# for loop to read in each CSV file in the directory into a dataframe,
# drop unnecessary columns, and append it to a dataframe

for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        full_filename = (os.path.join(directory, filename))
        df = pd.read_csv(full_filename)
        df2 = df.drop(['BEGIN_YEARMONTH',
            'BEGIN_DAY',
            'BEGIN_TIME',
            'END_YEARMONTH',
            'END_DAY',
            'END_TIME',
            'EPISODE_ID',
            'CZ_TYPE',
            'CZ_FIPS',
            'CZ_NAME',
            'WFO',
            'BEGIN_DATE_TIME',
            'CZ_TIMEZONE',
            'END_DATE_TIME',
            'SOURCE',
            'MAGNITUDE_TYPE',
            'FLOOD_CAUSE',
            'TOR_OTHER_WFO',
            'TOR_OTHER_CZ_STATE',
            'TOR_OTHER_CZ_FIPS',
            'TOR_OTHER_CZ_NAME',
            'BEGIN_RANGE',
            'BEGIN_AZIMUTH',
            'BEGIN_LOCATION',
            'END_RANGE',
            'END_AZIMUTH',
            'END_LOCATION',
            'END_LAT',
            'END_LON',
            'EPISODE_NARRATIVE',
            'EVENT_NARRATIVE',
            'DATA_SOURCE'], axis=1)
        # store DataFrame in list
        appended_data.append(df2)
        continue
    else:
        continue
appended_data = pd.concat(appended_data)
```

```python
# populate NULL values with 0
appended_data = appended_data.fillna(0)

# convert STATE_FIPS from FLOAT to INT
appended_data['STATE_FIPS'] = appended_data['STATE_FIPS'].astype(int)

# remove data after 2020 so that in lines up with CO2 data date range
appended_data = appended_data.drop(appended_data[appended_data.YEAR > 2020].index)
```

```python
# export data to CSV file
appended_data.to_csv(os.path.join(directory,'SevereWeatherDetails_appended.csv'), index=False)
```
