```python
# import dependencies
import pandas as pd
import os
```

```python
# read csv file into dataframe
df = pd.read_csv('co2 data by country by year.txt')
df.head(20)
```

```python
# group all data by year and add up values
df2 = df.groupby('year', as_index=False).sum()
df2.head(20)
```

```python
df2.tail(20)
```

```python
# drop all data older than 1950
df3 = df2.drop(df2[df2.year < 1950].index)
df3.head(20)
```

```python
df3.columns
```

```python
# drop unused columns 
df4 = df3.drop(['consumption_co2',
    'co2_growth_prct',
    'co2_growth_abs',
    'trade_co2',
    'consumption_co2_per_capita',
    'share_global_co2',
    'cumulative_co2',
    'share_global_cumulative_co2',
    'co2_per_gdp',
    'consumption_co2_per_gdp',
    'cement_co2_per_capita',
    'coal_co2_per_capita',
    'flaring_co2_per_capita',
    'gas_co2_per_capita',
    'oil_co2_per_capita',
    'other_co2_per_capita',
    'trade_co2_share',
    'share_global_cement_co2',
    'share_global_coal_co2',
    'share_global_flaring_co2',
    'share_global_gas_co2',
    'share_global_oil_co2',
    'share_global_other_co2',
    'cumulative_cement_co2',
    'cumulative_coal_co2',
    'cumulative_flaring_co2',
    'cumulative_gas_co2',
    'cumulative_oil_co2',
    'cumulative_other_co2',
    'share_global_cumulative_cement_co2',
    'share_global_cumulative_coal_co2',
    'share_global_cumulative_flaring_co2',
    'share_global_cumulative_gas_co2',
    'share_global_cumulative_oil_co2',
    'share_global_cumulative_other_co2'], axis=1)
df4.head(20)
```

```python
# export dataframe to csv file
df4.to_csv(('co2_byYear.csv'), index=False)
```
